```python
countyPops=pd.read_csv('co-est2021-alldata.csv')
print(countyPops.shape)
print(countyPops.columns)
def zeroPad(anInt):
    if (anInt < 10):
        return '00' + str(anInt)
    elif (anInt < 100):
        return '0' + str(anInt)
    else:
        return str(anInt)
    
def geoID(state,county):
    return str(state) + zeroPad(int(county))
    
countyPops['GEOID']=countyPops.apply(lambda df: geoID(df.STATE,df.COUNTY),axis=1)
#print(countyPops[['STATE','COUNTY','GEOID']])
popDct=dict(zip(countyPops['GEOID'].values,countyPops['POPESTIMATE2020'].values))

statePopDF=countyPops[countyPops['COUNTY']==0]

statePopDct=dict(zip(statePopDF['STNAME'].values,statePopDF['POPESTIMATE2020'].values))
print(statePopDct)
print(sum([v for v in statePopDct.values()]))
stateDct=dict(zip(countyPops['GEOID'].values,countyPops['STNAME'].values))
#set(countyPops['STNAME'].values)

#print(countyPops)
```

```python
import numpy

plural='Temperatures'
singular='Temperature'

#plural='Humidities'
#singular='Humidity'
overallDFrame=pd.read_csv('all%sByCounty.csv'%plural)
#print(overallDFrame.shape)
#print(overallDFrame.columns)
#print(overallDFrame[overallDFrame[singular]==-1]['County'])
overallDFrame=overallDFrame[overallDFrame[singular] != -1]

#print(overallDFrame.shape)
print(overallDFrame['Date'].values[0:100])
```

```python
def countyCodeToState(countyCode):
    if (int(countyCode) >= 72000):
        return 'Puerto Rico'
    #if (int(countyCode)==9009):
    #    print(countyCode,stateDct[countyCode])
    return stateDct[str(countyCode)]

def countyCodeToCountyPop(countyCode):
    #not using PR in analysis
    if (int(countyCode) >= 72000):
        return 0
    #if (int(countyCode)==9009):
    #    print(countyCode,stateDct[countyCode])
    return popDct[str(countyCode)]

overallDFrame['State']=overallDFrame['County'].apply(countyCodeToState)
overallDFrame['CountyPop']=overallDFrame['County'].apply(countyCodeToCountyPop)

stateToCounties=dict()
for i in range(overallDFrame.shape[0]):
    stateToCounties[overallDFrame['State'].values[i]]=stateToCounties.get(overallDFrame['State'].values[i],[]) + [overallDFrame['County'].values[i]]

for st in stateToCounties:
    stateToCounties[st]=list(set(stateToCounties[st]))
    
statePopDct=dict()

for st in stateToCounties:
    if (st=='Puerto Rico'):
        continue
    statePopDct[st]=numpy.sum([popDct[str(c)] for c in stateToCounties[st]])
    
statePopDct['Puerto Rico']=3194000

def getStatePop(st):
    return statePopDct[st]

overallDFrame['StatePop']=overallDFrame['State'].apply(getStatePop)

overallDFrame['StatePopShare']=overallDFrame['CountyPop']/overallDFrame['StatePop']

stateToNumCounties=dict()
for st in stateToCounties:
    stateToNumCounties[st]=len(set(stateToCounties[st]))
    
print(stateToNumCounties)

def getNumCounties(st):
    return stateToNumCounties[st]

overallDFrame['NumCountiesInState']=overallDFrame['State'].apply(getNumCounties)

print(statePopDct)
def monthAndYear(date):
    if (plural=='Humidities'):
        splts=date.strip().split('/')
        return (int(splts[0]),int(splts[2]))
    else:
        splts=date.strip().split('-')
        return (int(splts[1]),int(splts[0]))
        
overallDFrame['MonthAndYear']=overallDFrame['Date'].apply(monthAndYear)
#print(overallDFrame[overallDFrame.County==9009]['County'])
#countyCodeToState(9009)
#overallDFrame['RelativeStatePopShare']

overallDFrame['RelativeStatePopShare']=overallDFrame['StatePopShare']*overallDFrame['NumCountiesInState']
overallDFrame[singular + 'TimesRelativeStatePopShare']=overallDFrame[singular]*overallDFrame['RelativeStatePopShare']

overallDFrame[singular + 'TimesStatePopShare']=overallDFrame[singular]*overallDFrame['StatePopShare']
```
