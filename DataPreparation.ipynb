The MobilityResidual2MonthsPrev variable is NOT used for the results presented in the paper. It is only used as part of an alternate specification whose results are briefly mentioned in the supplemental.Results using the mobility residual are only presented for the sake of satisfying the curiosity of those readers who might be tempted to draw inferences when regarding the coefficient for the Oxford stringency index. The stringency index is likely mediated by the Google mobility variable. Including the both as controls as-is suffices for the sake of studying relative humidity, but if the focus is on social distancing policy as measured by the stringency index, we can look at the approach taken in "Vickers D, Baral S, Mishra S, Maheu-Giroux M, Buckridge D, Williamson T. Stringency of containment and closures on the growth of SARS-CoV-2 in Canada prior to accelerated vaccine roll-out. International Journal of Infectious Diseases. Vol 118, 73-82, 2022". In that paper, mobility is regressed against the stringency index and the residual of that regression is controlled for. Doing so in this study does not affect the relative humidity coefficient but does affect the stringency index coefficient substantially.  

We emphasize that disagreement with this residualizing statistical methodology should not influence evaluation of the paper itself, since the residualizing methodology is not used anywhere in the main paper results (only in the supplemental).

```python
mobilityRegressedStringency=smf.ols('MobilityReduction2MonthsPrev~Stringency2MonthsPrev',data=panelDF).fit()

slope=mobilityRegressedStringency.params['Stringency2MonthsPrev']
print(slope,mobilityRegressedStringency.params['Intercept'])
panelDF['MobilityResidual2MonthsPrev']=panelDF['MobilityReduction2MonthsPrev'] -slope*panelDF['Stringency2MonthsPrev'] -mobilityRegressedStringency.params['Intercept']
#print(panelDF[panelDF['MonthAndYear']=='2021_3']['MobilityReduction2MonthsPrev'].median())
```

```python
medianHumidityByState=panelDF.groupby('State')['RelativeHumidity2MonthsPrev'].median()

def medianHumidity(state):
    return medianHumidityByState[state]

panelDF['MedianRelativeHumidity']=panelDF['State'].apply(medianHumidity)

panelDF['HumidityDeviation2MonthsPrev']=panelDF['RelativeHumidity2MonthsPrev']- panelDF['MedianRelativeHumidity']

panelDF['HumidityDeviation1MonthsPrev']=panelDF['RelativeHumidity1MonthsPrev']- panelDF['MedianRelativeHumidity']

panelDF['HumidityDeviation3MonthsPrev']=panelDF['RelativeHumidity3MonthsPrev']- panelDF['MedianRelativeHumidity']

panelDF['HumidityDeviationCurrent']=panelDF['RelativeHumidityCurrent']- panelDF['MedianRelativeHumidity']


meanHumidityByState=panelDF.groupby('State')['RelativeHumidity2MonthsPrev'].mean()

def meanHumidity(state):
    return meanHumidityByState[state]

panelDF['MeanRelativeHumidity']=panelDF['State'].apply(meanHumidity)

panelDF['HumidityDeviationFromMean2MonthsPrev']=panelDF['RelativeHumidity2MonthsPrev']- panelDF['MeanRelativeHumidity']

panelDF['HumidityDeviationFromMean1MonthsPrev']=panelDF['RelativeHumidity1MonthsPrev']- panelDF['MeanRelativeHumidity']

panelDF['HumidityDeviationFromMean3MonthsPrev']=panelDF['RelativeHumidity3MonthsPrev']- panelDF['MeanRelativeHumidity']

panelDF['HumidityDeviationFromMeanCurrent']=panelDF['RelativeHumidityCurrent']- panelDF['MeanRelativeHumidity']

for col in panelDF.columns:
    
    #print(col,'hhh')
    if ((col.find('Stringency') !=-1) or 
        (col.find('Mask') != -1) or (col.find('Boost') != -1) or (col.find('Cases') != -1) or 
        (col.find('Temperature') !=-1)  or (col.find('Death') != -1) or (col.find('Humidity') != -1) or 
       (col.find('Vax') != -1) or (col.find('Fahrenheit') != -1 ) or
        (col.find('Excess') != -1) or (col.find('Mobility') != -1)  ):
        mean=panelDF[col].mean()
        std=panelDF[col].std()
        panelDF[col + '_stdized']=(panelDF[col] - mean)/std

panelDF.to_csv('panelDataFrame.csv')
#print(sorted(panelDF.columns))
#print(mobilityRegressedStringency.params)
```
